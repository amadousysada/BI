```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
data = pd.read_csv("eurostat/eurostat-2013.csv")
```

```python
data.iloc[0]
```

```python
data['tsc00004 (2012)'] = data['tsc00004 (2012)'].apply(lambda x: x/data.loc[lambda df: df['tsc00004 (2012)']==38637]["tps00001 (2013)"])
```

```python
data =data.drop(['tps00001 (2013)'], axis=1)
```

```python
data
```

```python
print(np.std(data))
```

```python
standardscaler = preprocessing.StandardScaler()
```

```python
X = data.iloc[:,2:11]
#y = (0.3-np.mean(data['tec00115 (2013)']))/np.std(data['tec00115 (2013)'][0])
Y= X.columns
Y
```

```python
X_norm = standardscaler.fit_transform(X)
```

```python
X_norm
```

```python
pca = PCA()
```

```python
X_pca =pca.fit_transform(X_norm)
X_pca.shape
```

```python
Y.values.shape
```

```python

pca1_pca2 = pd.DataFrame(data=X_pca, columns = Y)
```

```python
finalDf = pd.concat([pca1_pca2, data[['Code']]], axis = 1)
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('facteur 1', fontsize = 15)
ax.set_ylabel('facteur 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = data['Code'].values
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Code'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'teilm (M dec 2013)']
               , finalDf.loc[indicesToKeep, 'tec00118 (2013)']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
